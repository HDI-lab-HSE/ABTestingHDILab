# Генерация датасета

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

## Функции 

```python
def gen_df(size:int, group = True, csv = False):
    
    #size - требуемый размер датафрейма, 
    #group - флаг, нужно ли разбиение на группы
    #csv - флаг нужно ли сохранение в csv
    
    CUSTOMER_TYPE = [] # тип клиента Юр. лицо или ИП
    ACC_BALANCE_RUR = [] # баланс на счете - числовая, дробная, 2 знака поле запятой, рубли и копейки.
    CHEP = [] # ЧЭП - чистая экономическая прибыль - числовая, целые числа
    REGISTRATION_DT = [] # как давно клиент зарегестрирован в днях, числовая, целые числа
    CRM_CA_TB_NM = [] # категориальная - терр. банк
    TOTAL_CNT = [] # целевая переменная - количество сделок, числовая, целые числа
    GROUP = [] # флаг группы
    INN = [] # номер участника

    CRM_CA_TB_NM_cat = ['Юго-Западный банк', 'Центральный аппарат', 'Уральский банк',
     'Поволжский банк', 'Байкальский банк', 'Сибирский банк',
     'Северо-Западный банк', 'Волго-Вятский банк', 'Среднерусский банк',
     'Московский банк', 'Центрально-Черноземный банк', 'Дальневосточный банк']

    CUSTOMER_TYPE_cat = ['Юр. лицо', 'ИП']


    for i in range(size):
    
        CUSTOMER_TYPE.append(random.choice(CUSTOMER_TYPE_cat))
        CRM_CA_TB_NM.append(random.choice(CRM_CA_TB_NM_cat))
        ACC_BALANCE_RUR.append(round(random.uniform(84442,5707114),2))     
        CHEP.append(random.randint(1127,30359))
        REGISTRATION_DT.append(random.randint(2354,6660))
        TOTAL_CNT.append(random.randint(0,25))        
        INN.append(i)
        
        if(group == True):
            GROUP.append(random.randint(0,1))
                
    
    data = {'INN': INN, 'CUSTOMER_TYPE': CUSTOMER_TYPE, 'ACC_BALANCE_RUR': ACC_BALANCE_RUR, 'REGISTRATION_DT': REGISTRATION_DT, 'CRM_CA_TB_NM':CRM_CA_TB_NM, 'TOTAL_CNT':TOTAL_CNT} 
    df = pd.DataFrame(data) 
    
    
    if(group == True):
            df['GROUP'] = GROUP
                
    if(csv == True):
        
        df = df.set_index('INN')
        df.to_csv('bank_data_generation.csv', sep=';', encoding='utf-8-sig')
        df.reset_index(inplace= True)
        

    return df
```

```python
# пример использования
gen_df(size=1000, group=True, csv=True)
```
